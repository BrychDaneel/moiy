```python
import numpy
from numpy import linalg

from sympy.solvers import solve
from sympy import Symbol, Eq
```

```python
def quad(a,b,c,d,x,jop,jhop):
    while True:
        print("IT")
        n = len(x)
        m = len(a)
        aop = a[:, jop]
        naop = linalg.inv(aop)
        njop = set(range(n)) - set(jop)

        fc = (c.T + x.T * d).T
        fu = (- fc[jop].T * naop).T
        fd = (fu.T * a + fc.T).T
        print(fd)

        if (fd >= -0.0000001).all():
            return x

        j0 = 0
        while fd[j0] >= 0:
            j0 += 1
            print("adsasd")

        l = numpy.array([0] * n, dtype=numpy.float64)
        for j in njop:
            if j == j0:
                l[j] = 1
            else:
                l[j] = 0

        hul = d[:, jhop][jhop, :]
        hdl = a[:, jhop]
        hl = numpy.concatenate((hul, hdl))
        
        hur = a[:, jhop].T
        hz = numpy.matrix(numpy.zeros((len(hl) - len(hur), hur.shape[1]), dtype=numpy.float64))
        hr = numpy.concatenate((hur, hz))
    
        h = numpy.concatenate((hl, hr), 1)
        nh = linalg.inv(h)     

        bs = a[:,j0]
        badd = d[jhop, :][:, j0]
        bs = - numpy.concatenate((badd, bs))
        xy = nh * bs
        l[jhop] = xy[ :len(jhop), :].transpose()
        l = numpy.matrix(l).T
        et = l.T * d * l
        qj0 = numpy.Infinity if et == 0 else abs(fd[j0]) / et


        q0 = qj0
        js = j0
        for j in jhop:
            q = numpy.Infinity if l[j] >= 0 else - x[j] / l[j]
            if q < q0:
                q0 = q
                js = j

        if q0 == numpy.Infinity:
            print('Неограниченно снизу')
            return None

        x = x + q0[0,0] * l

        if j0 == js:
            jhop.append(js)
            continue

        if js in jhop and js not in jop:
            jhop.remove(js)
            continue

        s = jop.index(js)
        jp = None
        for j in set(jhop) - set(jop):
            if (naop * a[:, jp])[s] != 0:
                jp = j

        if jp is not None:
            jop[s] = jp
            jhop.remove(js)
        else:
            jop[jop.index(js)] = j0
            jhop[jhop.index(js)] = j0
```
