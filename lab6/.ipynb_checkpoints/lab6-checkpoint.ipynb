```python
import numpy
from numpy import linalg

from sympy.solvers import solve
from sympy import Symbol, Eq
```

```python
def quad(a,b,c,d,x,jop,jhop):
    while True:
        print('===========================')
        print('a: ', a)
        print('x: ', x)
        print('jop: ',jop)
        print('jhop', jhop)
        n = len(x)
        m = len(a)
        aop = a[:, jop]
        naop = linalg.inv(aop)
        njop = set(range(n)) - set(jop)

        fc = (c.T + x.T * d).T
        fu = (- fc[jop].T * naop).T
        fd = (fu.T * a + fc.T).T

        if (fd >= -0.0000001).all():
            return x

        j0 = 0
        while fd[j0] >= 0:
            j0 += 1
        
        print('j0', j0)

        l = numpy.array([0] * n, dtype=numpy.float64)
        for j in njop:
            if j == j0:
                l[j] = 1
            else:
                l[j] = 0

        hul = d[:, jhop][jhop, :]
        hdl = a[:, jhop]
        hl = numpy.concatenate((hul, hdl))
        
        hur = a[:, jhop].T
        hz = numpy.matrix(numpy.zeros((len(hl) - len(hur), hur.shape[1]), dtype=numpy.float64))
        hr = numpy.concatenate((hur, hz))
    
        h = numpy.concatenate((hl, hr), 1)
        print('h:')
        print(h)
        nh = linalg.inv(h)     

        bs = a[:,j0]
        print('bu',bs)
        badd = d[jhop, :][:, j0]
        print('badd',bs)
        bs = -numpy.concatenate((badd, bs))
        xy = nh * bs
        print('bs',bs)
        print('xy',xy)
        l[jop] = xy[:, :len(jhop)]
        l = numpy.matrix(l).T
        et = l.T * d * l
        qj0 = numpy.Infinity if et == 0 else abs(fd[j0]) / et
        
        print('l:', l)

        q0 = qj0
        js = j0
        print('q: ',q0[0][0])
        for j in jhop:
            q = numpy.Infinity if l[j] >= 0 else - x[j] / l[j]
            print('q, j, lj: ',q,j,l[j])
            if q < q0:
                q0 = q
                js = j

        if q0 == numpy.Infinity:
            print('Неограниченно снизу')
            return None

        print('q0',q0[0,0])
        x = x + q0[0,0] * l

        if j0 == js:
            jhop.append(js)
            print('p1')
            continue

        if js in jhop and js not in jop:
            jhop.remove(js)
            print('p2')
            continue

        s = jop.index(js)
        jp = None
        for j in set(jhop) - set(jop):
            if (naop * a[:, jp])[s] != 0:
                jp = j

        if jp is not None:
            jop[s] = jp
            jhop.remove(js)
            print('p3')
        else:
            print('p4')
            jop[jop.index(js)] = j0
            jhop[jhop.index(js)] = j0
```

```python
a = numpy.matrix([
    [1, 0, 2, 1],
    [0, 1, -1, 2],
], dtype=numpy.float64)

b = numpy.matrix([
    [2],
    [2],
], dtype=numpy.float64)

c = numpy.matrix([
    [-8],
    [-6],
    [-4],
    [-6],
], dtype=numpy.float64)

d = numpy.matrix([
    [2, 1, 1, 0],
    [1, 1, 0, 0],
    [1, 0, 1, 0],
    [0, 0, 0, 0],
], dtype=numpy.float64)

x = numpy.matrix([
    [2],
    [3],
    [0],
    [0]
], dtype=numpy.float64)

jop = [0, 1]
jhop = [0, 1]

quad(a, b, c, d, x, jop, jhop)
```
